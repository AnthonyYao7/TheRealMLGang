```python
import pandas as pd
import string
import re

# read data csv
data = pd.read_csv("dataset.csv")

print(data)
```

```python
# drop track_id column
#data = data.drop("track_id", axis=1)
#data = data.drop("album_name", axis = 1)
data = data.drop("track_genre", axis = 1)
data = data.drop(data.columns[0], axis=1)
data = data.drop("duration_ms", axis = 1)
data = data.drop("explicit", axis = 1)
data = data.drop("danceability", axis = 1)
data = data.drop("energy", axis = 1)
data = data.drop("key", axis = 1)
data = data.drop("loudness", axis = 1)
data = data.drop("mode", axis = 1)
data = data.drop("speechiness", axis = 1)
data = data.drop("acousticness", axis = 1)
data = data.drop("instrumentalness", axis = 1)
data = data.drop("liveness", axis = 1)
data = data.drop("valence", axis = 1)
data = data.drop("tempo", axis = 1)
data = data.drop("time_signature", axis = 1)
print(data)
```

```python

# drop rows with popularity 0
data = data[data['popularity'] != 0]

# fix the index of the resulting dataframe
data.reset_index(drop=True, inplace=True)
 
print(data)
```

```python
def contains_non_english(text):
    pattern = r'[^\x00-\x7F]+'
    return bool(re.search(pattern, text))

# Filter rows that contain non-English characters and drop them
data = data[~data['track_name'].apply(contains_non_english)].reset_index(drop=True)
print(data)
```

```python
data = data.sort_values("popularity")
data = data.drop_duplicates(keep='first')
top_data = data.head(10000)
top_data = top_data.drop("album_name", axis = 1)
top_data = top_data.drop("popularity", axis = 1)
print(top_data)
```
