```python
import lyricsgenius
import pandas as pd
import csv

artists = pd.read_csv("cleaned_data.csv")
genius = lyricsgenius.Genius("FZBr75i9sjMVy2UPPywD8_gQH8F2Vsk1Gd12GbrlN6yKp0Vbw7HpzmkDr7mfIbsH")

df = pd.DataFrame(artists)
```

```python
import time
from retry import retry

# Define the retry decorator
@retry(tries=5, delay=1)
def run_cell():
    # Check if a row already exists in the CSV file
    def is_duplicate_row(curr_artist, song_title):
        with open('lyrics.csv', 'r', encoding='utf-8') as csvfile_in:
            reader = csv.reader(csvfile_in)
            for row in reader:
                if row[0] == curr_artist and row[1] == song_title:
                    return True
        return False

    # Get the number of rows in the CSV file
    def get_num_rows():
        with open('lyrics.csv', 'r', encoding='utf-8') as csvfile_in:
            reader = csv.reader(csvfile_in)
            return sum(1 for row in reader)

    # Save a row to the CSV file
    def save_row(curr_artist, curr_track, lyrics, max_length):
        if len(lyrics) <= max_length:
            with open('lyrics.csv', 'a', encoding='utf-8', newline='') as csvfile_out:
                writer = csv.writer(csvfile_out)
                writer.writerow([curr_artist, curr_track, lyrics])

    # Resume from the last row that was written to the CSV file
    start_row = get_num_rows()

    # Iterate through the remaining rows in the dataframe
    for index, row in df.iloc[start_row:].iterrows():
        curr_track = row['track_name']
        curr_artist = row['artists']
        song = genius.search_song(curr_track, curr_artist)

        if song is None: 
            continue

        if not is_duplicate_row(curr_artist, song.title):
            save_row(curr_artist, curr_track, song.lyrics, max_length=10000)

# Call the retry decorator
run_cell()
```
